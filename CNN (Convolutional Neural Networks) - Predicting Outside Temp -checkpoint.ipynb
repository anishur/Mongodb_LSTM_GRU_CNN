```python
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.layers import Flatten
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from datetime import datetime
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('outside_temp.csv')
df.tail(10)
df.shape
```

```python
df['Timestamp'] = df.index
df.plot(x='Timestamp', y='temp_mean', kind='line', figsize=(20,6), rot=24)
```

### Preparing the Data

```python
FullData = df[['temp_mean']].values
print(FullData[0:5])
FullData.shape
```

```python
# normalization 

sc = MinMaxScaler()

DataScaler = sc.fit(FullData)

#print(DataScaler)

X = DataScaler.transform(FullData)

print('## after normalization ##')

X[0:5]

X.shape
```

### Preparing the data for CNN

```python
X_samples = list()
y_samples = list()

NumberOfRows = len(X)
TimeSteps = 10          # next days price prediction based on last how many past days values-- here 10 values


# Iterate through the values to create combinations

for i in range(TimeSteps, NumberOfRows, 1):
    x_sample = X[i-TimeSteps:i]
    y_sample = X[i]
    X_samples.append(x_sample)
    y_samples.append(y_sample)


# Reshape the Input as a 3D (number of samples, Time Steps, Features)

X_data = np.array(X_samples)
X_data = X_data.reshape(X_data.shape[0], X_data.shape[1], 1)
print('## Input Data shape ##')
print(X_data.shape)


# we do not need to reshape y as a 3D data as it is suppose to be a single column only

y_data = np.array(y_samples)
y_data = y_data.reshape(y_data.shape[0], 1)
print('## Output Data Shape ##')
print(y_data.shape)


```

```python
# Getting data for CNN
# Splitting the data into training and testing 
# Keeping last few days of data to test the learnings of the model and rest for training the model
# last 5 days as testing
```

```python
# Choosing the number of testing data records

TestingRecords = 5

# Splitting the data into train and test 

X_train = X_data[:-TestingRecords]
X_test  = X_data[-TestingRecords:]
y_train = y_data[:-TestingRecords]
y_test  = y_data[-TestingRecords:]

# Printing the shape of training and testing 

print('### Training Data Shape ###')
print(X_train.shape)
print(y_train.shape)

print('### Testing Data Shape ###')
print(X_test.shape)
print(y_test.shape)
```

```python
# Creating the Deep Learning CNN model

# Defining the Input shapes for CNN

TimeSteps = X_train.shape[1]
TotalFeatures = X_train.shape[2]
print("Number of TimeSteps:", TimeSteps)
print("Number of Features:", TotalFeatures)
```

#### Initialise, compile and fit the CNN


```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(TimeSteps, TotalFeatures)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
# Fitting the CNN to the Training set 

StartTime = time.time()

model.fit(X_train, y_train, epochs=500, batch_size=150)

EndTime = time.time()

print("## Total Time Taken: ", round((EndTime-StartTime)/60), 'Minutes ##')
```

```python
# Making the predictions on test data

predicted_Value = model.predict(X_test)
predicted_Value = DataScaler.inverse_transform(predicted_Value)


# Getting the Original value for testing data

orig = y_test
orig = DataScaler.inverse_transform(y_test)

# Accuracy of the predictions 

print('Accuracy: ', 100 - (100 * (abs(orig-predicted_Value)/orig)).mean())
```

```python
# Visualizing the results

plt.plot(predicted_Value, color='blue', label = 'Predicted Values')
plt.plot(orig, color='green', label ='Original Values')

plt.title('Outside Temp value Prediction using CNN')
plt.xlabel('Date')
plt.xticks(range(TestingRecords), df.tail(TestingRecords)['Timestamp'])
plt.ylabel('Temp ºC')

plt.legend()
fig=plt.gcf()
fig.set_figwidth(20)
fig.set_figheight(6)
plt.show()
```
