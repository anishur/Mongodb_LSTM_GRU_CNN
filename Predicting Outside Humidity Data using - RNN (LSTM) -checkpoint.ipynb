```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from datetime import datetime
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import GRU
import time
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df = pd.read_csv('outside_humidity.csv')
df.tail(10)
#df.shape
```

```python
df['Timestamp'] = df.index
df.plot(x='Timestamp', y='humidity_mean', kind='line', figsize=(20,6), rot=24)
```

#### Preparing the Data

```python
FullData = df[['humidity_mean']].values
print(FullData[0:5])
FullData.shape
```

```python
# normalization 

sc = MinMaxScaler()

DataScaler = sc.fit(FullData)

#print(DataScaler)

X = DataScaler.transform(FullData)

print('## after normalization ##')

X[0:5]
#X.shape
```

```python
# Preparing the data for LSTM

X_samples = list()
y_samples = list()

NumberOfRows = len(X)
TimeSteps = 10          # next days price prediction based on last how many past days values-- here 10 values


# Iterate through the values to create combinations

for i in range(TimeSteps, NumberOfRows, 1):
    x_sample = X[i-TimeSteps:i]
    y_sample = X[i]
    X_samples.append(x_sample)
    y_samples.append(y_sample)


# Reshape the Input as a 3D (number of samples, Time Steps, Features)

X_data = np.array(X_samples)
X_data = X_data.reshape(X_data.shape[0], X_data.shape[1], 1)
print('## Input Data shape ##')
print(X_data.shape)


# we do not need to reshape y as a 3D data as it is suppose to be a single column only

y_data = np.array(y_samples)
y_data = y_data.reshape(y_data.shape[0], 1)
print('## Output Data Shape ##')
print(y_data.shape)


```

```python
# Getting data for LSTM 
# Splitting the data into training and testing 
# Keeping last few days of data to test the learnings of the model and rest for training the model
# last 5 days as testing
```

```python
# Choosing the number of testing data records

TestingRecords = 5

# Splitting the data into train and test 

X_train = X_data[:-TestingRecords]
X_test  = X_data[-TestingRecords:]
y_train = y_data[:-TestingRecords]
y_test  = y_data[-TestingRecords:]

# Printing the shape of training and testing 

print('### Training Data Shape ###')
print(X_train.shape)
print(y_train.shape)

print('### Testing Data Shape ###')
print(X_test.shape)
print(y_test.shape)
```

```python
# Visualizing the input and output data for LSTM
# Printing some sample input and output values to help us visualize how the LSTM model will learn the values
# we can see the input is a 3D array of the last 10 days values 
# output is a 1D array of the next value
```

```python
# Visualizing the input and output being sent to the LSTM model

for inp, out in zip(X_train[0:2], y_train[0:2]):
    print(inp,'--',out)
```

```python
# Creating the Deep Learning LSTM model


# Defining the Input shapes for LSTM

TimeSteps = X_train.shape[1]
TotalFeatures = X_train.shape[2]
print("Number of TimeSteps:", TimeSteps)
print("Number of Features:", TotalFeatures)
```

```python
# Three hidden layers and one output layer is used for LSTM
```

```python
# Initialise the RNN(LSTM)

regressor = Sequential()

# Adding the first input layer

regressor.add(LSTM(units=10, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))
#regressor.add(GRU(units=10, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))

# Adding the second hidden layer

regressor.add(LSTM(units=5, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))
#regressor.add(GRU(units=5, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))

# Adding the third hiddeen layer

regressor.add(LSTM(units=10, activation = 'relu', return_sequences = False))
#regressor.add(GRU(units=10, activation = 'relu', return_sequences = False))

# Adding output layer

regressor.add(Dense(units=1))

# compiling the RNN (LSTM)

regressor.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Fitting the RNN to the Training set 

StartTime = time.time()

regressor.fit(X_train, y_train, batch_size= 5, epochs=100)

EndTime = time.time()

print("## Total Time Taken: ", round((EndTime-StartTime)/60), 'Minutes ##')
```

```python
# Making the predictions on test data

predicted_Value = regressor.predict(X_test)
predicted_Value = DataScaler.inverse_transform(predicted_Value)


# Getting the Original value for testing data

orig = y_test
orig = DataScaler.inverse_transform(y_test)

# Accuracy of the predictions 

print('Accuracy: ', 100 - (100 * (abs(orig-predicted_Value)/orig)).mean())
```

```python
# Visualizing the results

plt.plot(predicted_Value, color='blue', label = 'Predicted Volume')
plt.plot(orig, color='green', label ='Original Volume')

plt.title('Outside humidity value Prediction')
plt.xlabel('Date')
plt.xticks(range(TestingRecords), df.tail(TestingRecords)['Timestamp'])
plt.ylabel('Humidity in %')

plt.legend()
fig=plt.gcf()
fig.set_figwidth(20)
fig.set_figheight(6)
plt.show()
```

```python
# Generating predictions on full data

TrainPredictions = DataScaler.inverse_transform(regressor.predict(X_train))
TestPredictions = DataScaler.inverse_transform(regressor.predict(X_test))

FullDataPRedictions = np.append(TrainPredictions, TestingRecords)
FullDataOrig = FullData[TimeSteps:]

# Plot the full data

plt.plot(FullDataPRedictions, color='blue', label = 'Predicted Value')
plt.plot(FullDataOrig, color='green', label = 'Original Value')

plt.title('Humidity Value Predictions')
plt.xlabel('Date')
plt.ylabel('Humidity in %')
plt.legend()
fig=plt.gcf()
fig.set_figwidth(20)
fig.set_figheight(8)
plt.show()
```
