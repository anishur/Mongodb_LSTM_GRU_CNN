```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from datetime import datetime
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import GRU
import time
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df = pd.read_csv('outside_temp.csv')
df.tail(10)
#df.shape
```

```python
df['Timestamp'] = df.index
df.plot(x='Timestamp', y='temp_mean', kind='line', figsize=(20,6), rot=24)
```

### Preparing the Data

```python
FullData = df[['temp_mean']].values
print(FullData[0:5])
FullData.shape
```

```python
# normalization 

sc = MinMaxScaler()

DataScaler = sc.fit(FullData)

#print(DataScaler)

X = DataScaler.transform(FullData)

print('## after normalization ##')

X[0:5]
#X.shape
```

```python
# Preparing the data for LSTM

X_samples = list()
y_samples = list()

NumberOfRows = len(X)
TimeSteps = 10          # next days price prediction based on last how many past days values-- here 10 values


# Iterate through the values to create combinations

for i in range(TimeSteps, NumberOfRows, 1):
    x_sample = X[i-TimeSteps:i]
    y_sample = X[i]
    X_samples.append(x_sample)
    y_samples.append(y_sample)

#print(X_samples)
# Reshape the Input as a 3D (number of samples, Time Steps, Features)

X_data = np.array(X_samples)
#print(X_data)
X_data = X_data.reshape(X_data.shape[0], X_data.shape[1], 1)
print('## Input Data shape ##')
print(X_data.shape)


# we do not need to reshape y as a 3D data as it is suppose to be a single column only

y_data = np.array(y_samples)
y_data = y_data.reshape(y_data.shape[0], 1)
print('## Output Data Shape ##')
print(y_data.shape)


```

```python
# Getting data for LSTM 
# Splitting the data into training and testing 
# Keeping last few days of data to test the learnings of the model and rest for training the model
# last 5 days as testing
```

```python
# Choosing the number of testing data records

TestingRecords = 5

# Splitting the data into train and test 

X_train = X_data[:-TestingRecords]
X_test  = X_data[-TestingRecords:]
y_train = y_data[:-TestingRecords]
y_test  = y_data[-TestingRecords:]

# Printing the shape of training and testing 

print('### Training Data Shape ###')
print(X_train.shape)
print(y_train.shape)

print('### Testing Data Shape ###')
print(X_test.shape)
print(y_test.shape)
```

```python
# Visualizing the input and output data for LSTM
# Printing some sample input and output values to help us visualize how the LSTM model will learn the values
# we can see the input is a 3D array of the last 10 days values 
# output is a 1D array of the next value
```

```python
# Visualizing the input and output being sent to the LSTM model

for inp, out in zip(X_train[0:2], y_train[0:2]):
    print(inp,'--',out)
```

```python
# Creating the Deep Learning LSTM model


# Defining the Input shapes for LSTM

TimeSteps = X_train.shape[1]
TotalFeatures = X_train.shape[2]
print("Number of TimeSteps:", TimeSteps)
print("Number of Features:", TotalFeatures)
```

```python
# Three hidden layers and one output layer is used for LSTM
```

```python
#units=10: This means we are creating a layer with ten neurons 
#input_shape = (TimeSteps, TotalFeatures): The input expected by LSTM is in 3D format
#kernel_initializer=uniform: When the Neurons start their computation, some algorithm has to decide the value for each weight. This parameter specifies that. You can choose different values for it like normal or glorot_uniform.
#activation=relu: 
#return_sequences=True: LSTMs backpropagate thru time, hence they return the values of the output from each time step to the next hidden layer. This keeps the expected input of the next hidden layer in the 3D format. This parameter is False for the last hidden layer because now it does not have to return a 3D output to the final Dense layer.
#optimizer=adam: This parameter helps to find the optimum values of each weight in the neural network. adam is one of the most useful optimizers, another one is rmsprop
#batch_size=10: This specifies how many rows will be passed to the Network in one go after which the SSE calculation will begin and the neural network will start adjusting its weights based on the errors.
#When all the rows are passed in the batches of 10 rows each as specified in this parameter, then we call that 1-epoch. Or one full data cycle. This is also known as mini-batch gradient descent. A small value of batch_size will make the LSTM look at the data slowly, like 2 rows at a time or 4 rows at a time which could lead to overfitting, as compared to a large value like 20 or 50 rows at a time, which will make the LSTM look at the data fast which could lead to underfitting. Hence a proper value must be chosen using hyperparameter tuning.
#Epochs=10: The same activity of adjusting weights continues for 10 times, as specified by this parameter. In simple terms, the LSTM looks at the full training data 10 times and adjusts its weights.
```

```python
# Initialise the RNN(LSTM)

regressor = Sequential()

# Adding the first input layer

regressor.add(LSTM(units=10, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))
#regressor.add(GRU(units=10, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))

# Adding the second hidden layer

regressor.add(LSTM(units=5, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))
#regressor.add(GRU(units=5, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences = True))

# Adding the third hiddeen layer

regressor.add(LSTM(units=10, activation = 'relu', return_sequences = False))
#regressor.add(GRU(units=10, activation = 'relu', return_sequences = False))

# Adding output layer

regressor.add(Dense(units=1))

# compiling the RNN (LSTM)

regressor.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Fitting the RNN to the Training set 

StartTime = time.time()

regressor.fit(X_train, y_train, batch_size= 5, epochs=100)

EndTime = time.time()

print("## Total Time Taken: ", round((EndTime-StartTime)/60), 'Minutes ##')
```

```python
# Making the predictions on test data

predicted_Value = regressor.predict(X_test)
predicted_Value = DataScaler.inverse_transform(predicted_Value)


# Getting the Original value for testing data

orig = y_test
orig = DataScaler.inverse_transform(y_test)

# Accuracy of the predictions 

print('Accuracy: ', 100 - (100 * (abs(orig-predicted_Value)/orig)).mean())
```

```python
# Visualizing the results

plt.plot(predicted_Value, color='blue', label = 'Predicted Values')
plt.plot(orig, color='green', label ='Original Values')

plt.title('Outside Temp value Prediction')
plt.xlabel('Date')
plt.xticks(range(TestingRecords), df.tail(TestingRecords)['Timestamp'])
plt.ylabel('Temp C')

plt.legend()
fig=plt.gcf()
fig.set_figwidth(20)
fig.set_figheight(6)
plt.show()
```

```python
# We want to predict Next 5 days Temp values 
# we will use last 10 days values -> get Next 5 days values

# Data preparation for Multi Step LSTM

print('Original Temp Values')
print(FullData[-10:])
print('##############')

X = X.reshape(X.shape[0],)
print(X[-10:])
```

```python
# Multi step data preparation

# split into samples

X_samples = list()
y_samples = list()

NumerOfRows = len(X)
#print(NumerOfRows)
TimeSteps = 10         # this is last 10 days temp values
FutureTimeSteps = 5    # this is future 5 days to predict


# Iteration through the values to create combinations

for i in range(TimeSteps, NumerOfRows-FutureTimeSteps, 1):
    x_sample = X[i-TimeSteps:i]
    y_sample = X[i:i+FutureTimeSteps]
    X_samples.append(x_sample)
    y_samples.append(y_sample)


# Reshape the Input as a 3D (samples, Time Steps, Features)

X_data = np.array(X_samples)
X_data = X_data.reshape(X_data.shape[0], X_data.shape[1], 1)

print('### Input Data Shape ###')
print(X_data.shape)

# we donot need to reshape y as a 3D data as it is suppose to be a single column only

y_data = np.array(y_samples)

print('### Output Data Shape ###')
print(y_data.shape)
```

```python
# Input / output shape for LSTM multi step model
# splitting the data into Training and Testing

# Choosing the number of testing data records

TestingRecords = 5

# Splitting the data into train and test

X_train = X_data[:-TestingRecords]
X_test = X_data[-TestingRecords:]
y_train = y_data[:-TestingRecords]
y_test = y_data[-TestingRecords:]


# Printing the shape of training and testing 

print('### Training Data shape ###')
print(X_train.shape)
print(y_train.shape)

print('### Testing Data shape ###')
print(X_test.shape)
print(y_test.shape)
```

```python
# just to visualize the input --> output that set to LSTM Multi-step model
# Input vs. Output model for Multi step LSTM model

for inp, out in zip(X_train[0:2], y_train[0:2]):
    print(inp)
    print('===>')
    print(out)
```

```python
# Defining Input Shapes for LSTM

TimeSteps = X_train.shape[1]
TotalFeatures = X_train.shape[2]
print("Number of TimeSteps:", TimeSteps)
print("Number of Features:", TotalFeatures)
```

```python
# Initialize the RNN 

regressor = Sequential()

# Adding the first layer

regressor.add(LSTM(units=10, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences=True))
#regressor.add(GRU(units=10, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences=True))

# Adding the 2nd layer

regressor.add(LSTM(units=5, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences=True))
#regressor.add(GRU(units=5, activation = 'relu', input_shape = (TimeSteps, TotalFeatures), return_sequences=True))

# Adding the 3rd layer

regressor.add(LSTM(units=5, activation = 'relu', return_sequences=False))
#regressor.add(GRU(units=5, activation = 'relu', return_sequences=False))

# Adding the output layer

regressor.add(Dense(units= FutureTimeSteps))       # here we specify how many days to forcast / predict (FutureTimeSteps=5)
```

```python
# compile the RNN

regressor.compile(optimizer='adam', loss='mean_squared_error')

```

```python
# Fitting the RNN

StartTime = time.time()

regressor.fit(X_train, y_train, batch_size=5, epochs=100)

EndTime = time.time()

print('### Total Time Taken: ', round((EndTime-StartTime)/60), 'Minutes')
```

```python
# Measuring the accuracy of the model on testing data

# Making the predictions on test data

# Original and Predicted prices in LSTM Multi Step model

predicted_Value = regressor.predict(X_test)
predicted_Value = DataScaler.inverse_transform(predicted_Value)

print('### Predicted Temp Value ###')
print(predicted_Value)

# Getting the original values for testing data

orig = y_test
orig = DataScaler.inverse_transform(y_test)

print('### Original Temp Value ###')
print(orig)
```
